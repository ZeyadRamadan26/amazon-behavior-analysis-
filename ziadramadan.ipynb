```python
yesandno = {"Yes":1,"No":0}
yes_noandsometimes = {"Yes":1,"No":0,"Sometimes":2}
yes_noandmaybe = {"Yes":1,"No":0,"Maybe":3}
purchaselabels ={"Few times a month":0,"Once a week":1,"Once a month":2,"Multiple times a week":3,"Less than once a month":4}
gender = {"Male":1,"Female":0,"Prefer not to say":4 ,"Others": 3}
```

```python
df.head()
```

هستعمل label encoder بدل معملها manual لكل column

```python
df['Gender'] = df['Gender'].map(gender)
df['Personalized_Recommendation_Frequency'] = df['Personalized_Recommendation_Frequency'].map(yesandno)
df['Recommendation_Helpfulness'] = df['Recommendation_Helpfulness'].map(yes_noandsometimes)
df['Review_Left'] = df['Review_Left'].map(yesandno)
df['Review_Helpfulness'] = df['Review_Helpfulness'].map(yes_noandsometimes)
df['Purchase_Frequency'] = df['Purchase_Frequency'].map(purchaselabels)
```

```python
df.head()
```

```python
columns_to_encode = ['Gender', 'Personalized_Recommendation_Frequency',
                     'Recommendation_Helpfulness', 'Review_Left', 'Review_Helpfulness']
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
df['Add_to_Cart_Browsing'] = df['Add_to_Cart_Browsing'].map(yes_noandmaybe)
```

```python
df.head()
```

```python
lastcat_columns = df.select_dtypes(include=['object']).columns
print("Non-numeric columns:", lastcat_columns)
```

```python
for column in lastcat_columns:
    label_encoder = LabelEncoder()
    df[column] = label_encoder.fit_transform(df[column])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
sse = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
```

```python
from sklearn.preprocessing import StandardScaler
scalecolums = StandardScaler()
df_scaledcolumn = pd.DataFrame(scalecolums.fit_transform(df), columns=df.columns)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_values, sse, marker='o')
plt.title('Elbow Method for Optimal number of clusters')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of squared distances')
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, random_state=42)
df['KMeans_Cluster'] = kmeans.fit_predict(df)
```

```python
plt.figure(figsize=(12, 5))
plt.scatter(df_scaledcolumn["age"], df['Purchase_Frequency'], c=df['KMeans_Cluster'])
plt.title('K-Means Clustering')
plt.xlabel('age')
plt.ylabel('Purchase Frequency')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
plt.figure(figsize=(12, 5))
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['DBSCAN_Cluster'] = dbscan.fit_predict(df)
plt.scatter(df["age"], df['Purchase_Frequency'], c=df['DBSCAN_Cluster'])
plt.title('DBSCAN Clustering')
plt.xlabel('age')
plt.ylabel('Purchase Frequency')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(df, method='ward')
plt.figure(figsize=(12, 5))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('data')
plt.ylabel('Distance')
plt.show()
```
